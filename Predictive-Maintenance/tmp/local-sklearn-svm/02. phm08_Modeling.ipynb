```python
import pandas as pd
import numpy as np

train_df = pd.read_csv("../PHM2008/train_df.csv", delimiter=",",header="infer",index_col=None)
df = pd.read_csv("../PHM2008/train.csv", delimiter=",",header="infer",index_col=None)
```

```python
train_df["UnitNum"] = train_df["UnitNum"].astype(np.str)
df["UnitNum"] = df["UnitNum"].astype(np.str)
```

```python
train_df.head(10)
```

```python
num_col = ["ope1", "ope2", "ope3", "sensor1", "sensor2", "sensor3", "sensor4", "sensor5", "sensor6", "sensor7", "sensor8", "sensor9", "sensor10", "sensor11", "sensor12", "sensor13", "sensor14", "sensor15", "sensor16", "sensor17", "sensor18", "sensor19", "sensor20", "sensor21"]
```

```python
train_df[num_col] = train_df[num_col].apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
```

```python
train_df.describe()
```

```python
train_df.head(10)
```

```python
from sklearn import svm
```

```python
clf = svm.OneClassSVM(nu=0.01, kernel="rbf", gamma=1.0)
clf.fit(train_df.iloc[:,2:])
```

```python
df[num_col] = df[num_col].apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
```

```python
df.head(10)
```

```python
pred_train = clf.predict(df.iloc[:,2:])
```

```python
from scipy import stats

pred = clf.decision_function(df.iloc[:,2:]).ravel() # 各データの超平面との距離、ravel()で配列を1D化
threshold = stats.scoreatpercentile(pred, 100 * 0.1) # パーセンタイルで異常判定の閾値設定
```

```python
print(pred,threshold)
```

```python
df["pred"] = pd.DataFrame(pred,columns=["pred"])
```
